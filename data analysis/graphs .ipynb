```python
years = np.array([2019,2020,2021,2022,2023])
rainfall = np.array([377,489,421,542,594])
```

```python
# line plot
plt.plot(rainfall)
```

```python
plt.plot(years,rainfall)
```

```python
# configuring graph
plt.plot(years, rainfall)
plt.title("Rainfall har saal",fontdict={'fontsize':15,'color':'green'})
plt.xlabel("Years",color='yellow')
plt.ylabel("Rainfall(cm)",color='purple')
plt.xticks(years)
plt.hlines(y=500,xmin=2019,xmax=2023,color='red',linestyles='--')
plt.show()
```

```python
plt.rcParams['font.family'] = 'Copperplate Gothic Bold'
plt.figure(figsize=(5,3),facecolor='lightblue')
x = np.array(['Rahul','Vinay','Rohan','Rajesh','Raman'])
y = np.array([60,80,90,60,50])
bars = plt.bar(x,y,
        color = ['orange','green','violet','red','yellow'],
        alpha = .5, edgecolor = 'black',
        linewidth=1,hatch='/',width= .6)
bars[2].set_hatch('o')
bars[2].set_color('red')
plt.title("Parinam")
plt.xlabel("Bachchon ke naam")
plt.ylabel('Ank')
plt.ylim(0,120)
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2,
            bar.get_height()+10,
            f"${bar.get_height()}k",
            fontsize=10, color='red',
            ha = 'center')
plt.show()
```

```python
plt.figure(figsize=(5,5),facecolor='white')
plt.pie(
    y,
    labels = x,
    explode = [0,0,0.5,0,0],
    autopct='%.1f%%',
    pctdistance=.75,
    counterclock=True,
    startangle=90,
    wedgeprops={'width':.5,'edgecolor':'white'}, # donut chart
)
pie.show()
```
