```python
from PIL import Image, ImageDraw, ImageFont
import requests
```

```python
def download_image(url):
    response = requests.get(url,stream = True)
    return Image.open(response.raw)
```

```python
# online image
im1 = download_image('https://cdn.pixabay.com/photo/2021/12/26/17/31/pokemon-6895600_960_720.png')
im1
```

```python
# generated via code image
im3 = Image.new('RGB',(300,300),color='red')
im4 = Image.new('RGB',(300,300),color='green')
im3.show()
im4.show()
```

```python
# writing on image
im5 = Image.new('RGB',(300,300),color='green')
font = ImageFont.truetype('font.ttf',40)
font_sm = ImageFont.truetype('font.ttf',20)
draw = ImageDraw.Draw(im5)
draw.text(
    (150,150),
    text = 'Krashi',
    fill=(255,155,0),# yellow
    font=font,
    anchor='mm',# middle middle,
    allign = 'center'
)
draw.text((150,120),"Sanam",(255,255,255),font,'ms')
draw.text((150,280),"Together from PUBG",(0,0,0),font_sm,"mm")
im5.save('coded_image.png')
im5
```
